```python
df = pd.read_csv("/Users/yangshining/Desktop/internnnnn/IFC_summer_intern/IFC_intern_Shining/mapping_project/South Africa/Death/2015_deanths_raw.csv")
df.head()
```

```python
df["age_name_digits1"] = df.age_name.str.split('-',expand=True)[0]
df["age_name_digits2"] = df.age_name.str.split('-',expand=True)[1]
df["age_name_digits2"] = df.age_name_digits2.str.split(' ',expand=True)[0]
df["age_name_new"] = df["age_name_digits1"] + ' to ' + df["age_name_digits2"]
df["age_name"] = np.where(df["age_name"]== "80+ years", "80 plus", df["age_name_new"])
df["age_name"] = np.where(df["age_name"].isnull(), "<1 year", df["age_name"])
```
